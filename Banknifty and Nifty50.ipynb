```python
import pandas as pd
data = pd.read_csv("Banknifty2014_2018.csv", dtype={'Change in OI': int}, na_values=['-'])
data.rename(columns={'Symbol  ': 'Symbol', 'Date  ': 'Date','Expiry  ':'Expiry','Option type  ':'Option type','Strike Price  ':'Strike Price','Open  ':'Open','High  ':'High','Low  ':'Low','Close  ':'Close','LTP  ':'LTP','Settle Price  ':'Settle Price', 'No. of contracts  ':'No. of contracts','Turnover in Lakhs':'Turnover in Lakhs','Open Int  ':'Open Int','Change in OI  ':'Change in OI','Underlying Value  ':'Underlying Value'},inplace=True)
data
```

```python
#We remove the contracts that are not traded
data = data.dropna(subset=['Open'])
data = data.reset_index(drop=True)
data
```

```python
#Adding the close column of the previous day of the beginning of the contract
from datetime import datetime, timedelta
data.loc[:, 'Date'] = pd.to_datetime(data.loc[:, 'Date'])
selected_columns = ['Date', 'Expiry','Strike Price', 'Close']
extracted= data[selected_columns]
K_C=extracted.groupby('Date')
c1=[]
```

```python
#Get the data corresponding the day before and check if there is a contract with the same Expiry and Strike price, if it exists,
#We get the Close price corresponding to that index, else we assign 0.
for s in range(len(data)):
    start_date=data.loc[s,'Date']
    k=data.loc[s,'Strike Price']
    start_date = start_date - timedelta(days=1)
    if start_date in list(K_C.groups.keys()):    
        X=K_C.get_group(start_date)      #get the part of the groupped data for this date
        i = X['Strike Price'][(X['Strike Price'] == k) & (X['Expiry'] == data.loc[s, 'Expiry'])].index 
        if len(i)>0:         
            close=X.loc[i, 'Close']      #get the close price corresponding to that option contract
            c1.append((close.iloc[0])/k)  #add the value to the list c
        else:
            c1.append(0)
    else:
        c1.append(0)
```

```python
#Remove the rows that do are not traded on the previous day of the beginning of the contract
previous_day_close1 = pd.Series(c1, name='previous day close')
data['A day before_Close']=previous_day_close1
data=data[data['A day before_Close'] != 0]
data = data.reset_index(drop=True)
data
```

```python
def cov_statistics(cov_matrix):
    s=cov_matrix.shape
    for i in range(s[0]):
        for j in range(s[1]):
            if cov_matrix[i,j]==0:
                cov_matrix[i,j]=0
            else:
                cov_matrix[i,j]=(cov_matrix[i,j])/np.sqrt(abs(cov_matrix[i,j]))
    cov_matrix= cov_matrix[np.triu_indices(len(cov_matrix), k=0)]
    return(cov_matrix)
    
data['Covariance']=None
p=0
for i in range(len(data)):
    x=data.loc[i, 'Open_Log_returns']
    y=data.loc[i, 'High_Log_returns']
    v=data.loc[i, 'Low_Log_returns']
    w=data.loc[i, 'Close_Log_returns']
    data_matrix = np.vstack((x, y, v, w))
    covariance_matrix = np.cov(data_matrix)
    covar = cov_statistics(covariance_matrix)
    data.at[i, 'Covariance'] = covar

data
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
import tensorflow as tf
import numpy as np

# Encode the categorical labels using LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
num_classes = len(label_encoder.classes_)
y_onehot = to_categorical(y_encoded, num_classes=num_classes)

X_trainBI, X_testBI, y_trainBI, Y_testBI = train_test_split(X, y_onehot, test_size=0.3,shuffle=False)
train_indices = np.arange(len(X_trainBI))
test_indices = np.arange(len(X_trainBI), len(X_trainBI) + len(X_trainBI))

def precision_m(y_true, y_pred):
    true_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_true * y_pred, 0, 1)))
    predicted_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + tf.keras.backend.epsilon())
    return precision

def recall_m(y_true, y_pred):
    true_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_true * y_pred, 0, 1)))
    possible_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + tf.keras.backend.epsilon())
    return recall

BankApproachI = Sequential(name="feed_forward")
BankApproachI.add(Dense(128, activation='relu', input_shape=(X_testBI.shape[1],), name='Input-Layer'))
BankApproachI.add(Dense(64, activation='relu', name='Hidden-Layer'))
BankApproachI.add(Dense(num_classes, activation='softmax', name='Output-Layer'))

optimizer = Adam(learning_rate=0.00012)
BankApproachI.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy', precision_m])

BankApproachI.fit(X_trainBI,
          y_trainBI,
          batch_size=32,
          epochs=3,
          verbose=1,
          validation_split=0.2,
          shuffle=True,
          use_multiprocessing=False)



pred_probs_tr = BankApproachI.predict(X_trainBI)
pred_probs_te = BankApproachI.predict(X_testBI)

pred_labels_tr = pred_probs_tr.argmax(axis=1)
pred_labels_te = pred_probs_te.argmax(axis=1)

print("\n-------------------- Model Summary --------------------")
BankApproachI.summary()
print("\n-------------------- Weights and Biases --------------------")
for layer in BankApproachI.layers:
    print("Layer: ", layer.name)
    print("  --Kernels (Weights): ", layer.get_weights()[0])
    print("  --Biases: ", layer.get_weights()[1])

```

```python
def error_metricX(x,y):
    err=0
    T=len(x)
    for i in range(T):
        err+=0.1*((abs(x[i]-y[i]))/T)
    return(err)
def p_metricx(x,y):
    r=np.abs(x-y)
    count = np.sum(r > 2)
    return(count/len(x))
```

```python
#Get the data corresponding the day before and check if there is a contract with the same Expiry and Strike price, if it exists,
#We get the Close price corresponding to that index, else we assign 0.
for s in range(len(NIFTY50)):
    start_date=NIFTY50.loc[s,'Date']
    k=NIFTY50.loc[s,'Strike Price']
    start_date = start_date - timedelta(days=1)
    if start_date in list(K_C.groups.keys()):    
        X=K_C.get_group(start_date)      #get the part of the groupped data for this date
        i = X['Strike Price'][(X['Strike Price'] == k) & (X['Expiry'] == NIFTY50.loc[s, 'Expiry'])].index 
        if len(i)>0:         
            close=X.loc[i, 'Close']      #get the close price corresponding to that option contract
            c.append((close.iloc[0])/k)  #add the value to the list c
        else:
            c.append(0)
    else:
        c.append(0)
```

# Black–Scholes 

```python
def volatility(L):
    L=np.array(L)
    p=np.mean(L)
    r = [(x - p) / p for x in L]
    r = [x ** 2 for x in r]
    r=np.array(r)
    return(np.sqrt((1/19)*np.sum(r))*np.sqrt(252))

    
NIFTY50=NIFTY50.dropna(subset=['government bond yield rate'])   
NIFTY50 = NIFTY50.reset_index()
    
NIFTY50['Volatility']=None
p=0
for i in NIFTY50['last20_Close']:
    i = [float(value.replace(',', '')) for value in i]
    NIFTY50.at[p, 'Volatility'] = volatility(i)
    p+=1
```

```python
from scipy.stats import norm
NIFTY50['BS_price']=None
for i in range(len(NIFTY50)):
    T=NIFTY50.loc[i,'Time to maturiy'] / pd.to_timedelta(252, unit='D')
    sigma=NIFTY50.loc[i,'Volatility']
    K=NIFTY50.loc[i,'Strike Price']
    r=NIFTY50.loc[i,'government bond yield rate']/100
    S=NIFTY50.loc[i,'Underlying Value']
    d1=(np.log(S/K)+((r+(sigma**2)/2 )*T))/(sigma*np.sqrt(T))
    d2=d1-sigma*np.sqrt(T)
    Nd1=norm.cdf(d1)
    Nd2=norm.cdf(d2)
    PV=K*np.exp(-r*T)
    C=Nd1*S-Nd2*PV
    NIFTY50.at[i,'BS_price']=C
```

```python
def one_hot_to_real(predicted_onehot):
    pred_cat=[]
    r=predicted_onehot.shape[0]
    for i in range(r):
        pred_cat.append(np.argmax(predicted_onehot[i]))
    return(pred_cat)

def error_metric(x,y):
    err=0
    y=one_hot_to_real(y)
    x=one_hot_to_real(y)
    T=len(x)
    for i in range(T):
        err+=0.1*((abs(x[i]-y[i]))/T)
        print(i)
    return(err)
def p_metric(x,y):
    abs_diff = [abs(i - j) for i, j in zip(x, y)]
    count = len([diff for diff in abs_diff if diff > 2])
    return(count/len(x))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
import tensorflow as tf
import numpy as np

# Encode the categorical labels using LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
num_classes = len(label_encoder.classes_)
y_onehot = to_categorical(y_encoded, num_classes=num_classes)

X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.3,shuffle=False)
train_indices = np.arange(len(X_train))
test_indices = np.arange(len(X_train), len(X_train) + len(X_test))

def precision_m(y_true, y_pred):
    true_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_true * y_pred, 0, 1)))
    predicted_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + tf.keras.backend.epsilon())
    return precision

def recall_m(y_true, y_pred):
    true_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_true * y_pred, 0, 1)))
    possible_positives = tf.keras.backend.sum(tf.keras.backend.round(tf.keras.backend.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + tf.keras.backend.epsilon())
    return recall

NiftyApproachI = Sequential(name="feed_forward")
NiftyApproachI.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],), name='Input-Layer'))
NiftyApproachI.add(Dense(64, activation='relu', name='Hidden-Layer'))
NiftyApproachI.add(Dense(50, activation='softmax', name='Output-Layer'))

optimizer = Adam(learning_rate=0.000012)
NiftyApproachI.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
)

NiftyApproachI.fit(X_train,
          y_train,
          batch_size=32,
          epochs=3,
          validation_split=0.2,
          shuffle=False,
          use_multiprocessing=True)



pred_probs_tr = NiftyApproachI.predict(X_train)
pred_probs_te = NiftyApproachI.predict(X_test)

pred_labels_tr = pred_probs_tr.argmax(axis=1)
pred_labels_te = pred_probs_te.argmax(axis=1)

print("\n-------------------- Model Summary --------------------")
NiftyApproachI.summary()
print("\n-------------------- Weights and Biases --------------------")
for layer in NiftyApproachI.layers:
    print("Layer: ", layer.name)
    print("  --Kernels (Weights): ", layer.get_weights()[0])
    print("  --Biases: ", layer.get_weights()[1])

```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

# Specify the learning rate
learning_rate = 0.3

# Create the XGBoost DMatrix for training and testing data
dtrain = xgb.DMatrix(data=X_train, label=y_train)
dtest = xgb.DMatrix(data=X_test)

# Set the XGBoost parameters for multiclass classification
params = {
    'objective': 'multi:softmax',  # Multiclass classification objective
    'num_class': 64,     # Number of classes in your dataset
    'eval_metric': 'mlogloss',    # Multiclass logloss evaluation metric
    'learning_rate': learning_rate,
    'max_depth': 3,              # Maximum tree depth
}
# Specify the number of boosting rounds (equivalent to n_estimators)
num_boost_round = 100  

# Train the XGBoost model
model = xgb.train(params=params, dtrain=dtrain, num_boost_round=num_boost_round)
# Make predictions on the test data
predictions = model.predict(dtest)

# Convert the predictions to integers (class labels)
y_pred = predictions.astype(int)

# Calculate accuracy on the test set for multiclass classification
accuracy = accuracy_score(y_test, y_pred)
```
